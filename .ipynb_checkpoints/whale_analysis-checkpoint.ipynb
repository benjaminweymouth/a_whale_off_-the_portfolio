# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp_tsx_history.csv`: Contains historical closing prices of the S&P TSX 60 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data.

```python
whale_data = Path("./Resources/whale_returns.csv")
whale_data
```

```python
# Reading whale returns
whale_df = pd.read_csv(whale_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
whale_df.sort_index(axis=0, level=None, ascending=True, inplace=True, kind='quicksort')
whale_df.head()
 
```

```python
# Count nulls
whale_df.isnull().sum()
```

```python
# Drop nulls
whale_df.dropna(inplace=True)
#check for nulls 
whale_df.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data.

```python
# Reading algorithmic returns
algo_data = Path("./Resources/algo_returns.csv")
algo_data
```

```python
# Reading whale returns
algo_data_df = pd.read_csv(algo_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo_data_df.sort_index(axis=0, level=None, ascending=True, inplace=True, kind='quicksort')
algo_data_df.head() 
```

```python
# Count nulls
algo_data_df.isnull().sum()
```

```python
# Drop nulls
algo_data_df.dropna(inplace=True)
algo_data_df.head()
#check for nulls 
algo_data_df.isnull().sum()
```

```python
algo_data_df.head()
```

## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
sp_tsx_data = Path("./Resources/sp_tsx_history.csv")
sp_tsx_data
sp_tsx_df = pd.read_csv(sp_tsx_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp_tsx_df.sort_index(axis=0, level=None, ascending=True, inplace=True, kind='quicksort')
sp_tsx_df.head()
```

```python
# Check Data Types
sp_tsx_df.dtypes
```

```python
# Fix Data Types
sp_tsx_df["Close"] = sp_tsx_df["Close"].str.replace("$", "")
sp_tsx_df["Close"] = sp_tsx_df["Close"].str.replace(",", "")
sp_tsx_df.head()
```

```python
#convert to float 
sp_tsx_df["Close"] = sp_tsx_df["Close"].astype('float')

# Re-Check Data Types
sp_tsx_df["Close"].dtypes
sp_tsx_df.head()
```

```python
# Calculate Daily Returns
tsx_daily_returns = sp_tsx_df.pct_change()
tsx_daily_returns.head()
```

```python
# Count nulls
tsx_daily_returns.isnull().sum()

# Drop nulls

tsx_daily_returns.dropna(inplace=True)
tsx_daily_returns.head()
#check for nulls 
tsx_daily_returns.isnull().sum()
```

```python
# Rename `Close` Column to be specific to this portfolio.
tsx_daily_returns.columns = ["S&P TSX"]
tsx_daily_returns.head()
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
# Use the concat function to combine the DataFrames by column and perform an inner join. 
combined_df = pd.concat([whale_df, algo_data_df,tsx_daily_returns], axis="columns", join="inner")

# Sort datetime index in ascending order (past to present)
combined_df.sort_index(inplace=True)

# Display a few rows
combined_df.head()
```

## Risk Analysis

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios.
4. Determine which portfolios are riskier than the S&P TSX 60.
5. Calculate the Annualized Standard Deviation.

### Create a box plot for each portfolio


```python
# Box plot to visually show risk
combined_df.boxplot(figsize=(20,10))
```

### Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios
st_dev_series = combined_df.std()

#a series holding all the standard deviations
st_dev_series
```

### Determine which portfolios are riskier than the S&P TSX 60

```python
# Calculate  the daily standard deviation of S&P TSX 60
std_TSX = combined_df["S&P TSX"].std()

# Determine which portfolios are riskier than the S&P TSX 60
for index, value in st_dev_series.items():
        if index != "S&P TSX":
            if value > std_TSX:
                print(f"{index} has more risk than the S&P TSX 60 due to an std of {round(value, 5)}")
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_std = st_dev_series * np.sqrt(252)
annualized_std.head()
```

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
# Plot the rolling standard deviation
combined_df.rolling(window=21).std().plot(figsize=(15,10),title='Rolling standard deviation, Combined')
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation = combined_df.corr()

# Display de correlation matrix
correlation
```

```python
# Use the `heatmap` function from the Seaborn library to visualize the correlation table
sns.heatmap(correlation, vmin=-1, vmax=1,)
```

```python
# Alternatively, sum total correlation for each stock
correlation_series = correlation.sum(axis=0).sort_values()
correlation_series 
```

```python
#Find the 2 highest values indicating the highest correlation 

highest_value = correlation_series.max()
max_index = correlation_series.idxmax()
cleaned_series = correlation_series.drop(index=max_index)
second_highest_index = cleaned_series.idxmax()
second_highest_value = cleaned_series.max()
print(f"The two highest correlated stocks are {max_index} with {round(highest_value, 4)} and {second_highest_index} with {round(second_highest_value, 4)}")
```

```python
# Drop highly correlated stocks and keep only the non-correlated stocks
noncorrelated_daily_returns = combined_df.drop(columns=["SOROS FUND MANAGEMENT LLC", "Algo 2"])

# Display sample data
noncorrelated_daily_returns.head()
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio
chosen_covariance = combined_df['Algo 1'].cov(combined_df['S&P TSX'])
#print(chosen_covariance)

# Calculate variance of S&P TSX
 
tsx_variance = combined_df['S&P TSX'].var()
#print(variance)

# Computing beta
algo_beta = chosen_covariance / tsx_variance
#algo_beta


# Calculate 30-day rolling covariance of Algo 1 vs. TSX rolling variance
algo_rolling_covariance = combined_df['Algo 1'].rolling(window=30).cov(combined_df['S&P TSX'])
rolling_variance = combined_df['S&P TSX'].rolling(window=30).var()

# Plot beta trend
beta_trend = algo_rolling_covariance / rolling_variance
beta_trend.plot(figsize=(15,10), title='Rolling 30-Day Covariance of Algo 1 Returns vs. S&P TSX Returns')
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
ewm_result = combined_df.ewm(com=None, span=None, halflife=21).mean()
ewm_result.plot(figsize=(15,10), title='Exponentially Weighted Average')
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios

# Use the `mean` and `std` functions to calculate the annualized sharpe ratio
sharpe_ratios = (combined_df.mean() * 252) / (combined_df.std() * np.sqrt(252))

# Display sharpe ratios
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot.bar(x=None, y=None, figsize=(15,10), title='Sharpe Ratios')
```
